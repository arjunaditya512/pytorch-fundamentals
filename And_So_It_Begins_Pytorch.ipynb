```python
import torch
```

```python
scalar = torch.tensor(1) #scalars and 1D vectors in lower case
print(scalar)
print(scalar.item())
print(scalar.ndim)
```

```python
MATRIX = torch.tensor([[1,2],[3,4]]) #matrices and high dim tensors are wrtten in upper case
print(MATRIX)
print(MATRIX.ndim)
```

```python
print(MATRIX.shape)
```

```python
RANDOM = torch.rand(3,4,3)
print(RANDOM)
print(RANDOM.ndim)
```

```python
RAND_IMG_TENSOR = torch.rand(size=(224,224,3))
print(RAND_IMG_TENSOR)
print(RAND_IMG_TENSOR.ndim)
print(RAND_IMG_TENSOR.shape)
```

```python
ZEROES = torch.zeros(size=(3,4))
print(ZEROES)
print(ZEROES.ndim)

ONES = torch.ones(size=(3,4))
print(ONES)
print(ONES.ndim)
print(ONES.dtype)
```

```python
RANGE = torch.arange(start=0, end = 10, step=2)
print(RANGE)
```

```python
LIKE_RANGE = torch.zeros_like(input=RANGE)
print(LIKE_RANGE)
```

```python
#Tensor DATA TYPES
# 3 Errors in Torch:
# 1. Tensors not right datatype
# 2. Tensors not right shape
# 3. Tensors not on the right device

float32_type = torch.tensor(data =[1,2,3],
                         dtype=torch.float32,
                         device="cpu",    #cpu, cuda, NONE
                         requires_grad=False)
print(float32_type)
float_16= float32_type.type(torch.float16)
print(float_16)
```

```python
float_16 * float32_type
```

```python
%%time
A = torch.tensor([[1,2,3],[4,5,6]], dtype=torch.long)
B = torch.tensor([[7,8],[9,10],[11,12]], dtype=torch.long)
print(torch.matmul(A,B))
print(A.T@B.T)
```

```python
A = torch.tensor([1,2,3,4,5,6])
print(A.argmin(), A.argmax())
```

```python
#reshaping, stacking, squeezing, unsqueezing
A_reshape = A.reshape(3,2)
print(A_reshape)

A_reshape = A.view(2,3)
A_reshape

A_stack = torch.stack([A,A,A,A], dim=0)
print(A_stack)


A_stack = torch.stack([A,A,A,A], dim=1)
print(A_stack)

W = torch.tensor([[1,2,3],[4,5,6]])
W_stack = torch.stack([W,W,W,W], dim=2)
print(W_stack)
```

```python
sq = torch.tensor([[1,2,3,4,4]])
print(sq.shape)
print(sq.squeeze().shape)
# torch.tensor([[1,2,3,4,4]]).shape,torch.tensor([[1,2,3,4,4]]).squeeze().shape

unsq = torch.tensor([1,2,3,4,5])
print(unsq.shape)
print(unsq.unsqueeze(dim=0).shape)
print(unsq.unsqueeze(dim=1).shape)
```

```python
#combining torch with NUmpy

import numpy as np
arr = np.array([[1,2,3],[4,5,6]]) #default in float64
torch_tensor = torch.from_numpy(arr) #results in float64 tensor
print(arr)
print(torch.from_numpy(arr), torch_tensor.dtype)
print(torch_tensor.numpy())
#Changes in original np array will not change tensor and vice versa
```

```python
#saving tensors on CPU vs GPU
ts = torch.tensor([1,2,3,4,5], device='cpu')
#use .to_cpu() .to_gpu() to switch, numpy arrays can compute only on cpus
print(ts, ts.device)
```
